```python
x=df1.drop('charges',axis =1)
y=df['charges']
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
df1_pred = pd.DataFrame({'Actual': ytest, 'Prediction': pred, 'Difference': abs(ytest-pred)})
df1_pred.head()
```
