# Separeting x,y

```python
X = df.drop('charges',axis=1)
y = df['charges']
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
xtrain, xtest, ytrain, ytest = tts(X, y, test_size=0.3, random_state=78)
```

```python
reg = LinearRegression()
```

```python
reg.fit(xtrain,ytrain)
```

```python
lin_pred =reg.predict(xtest)
```

```python
lin_score =reg.score(xtest,ytest)
```

```python
lin_score
```

```python
m = reg.coef_
m
```

```python
c = reg.intercept_
c
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
lin_mae = mean_absolute_error(ytest, lin_pred)
lin_mae
```

```python
lin_mse = mean_squared_error(ytest, lin_pred)
lin_mse
```

# KNN 

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
xtrain, xtest, ytrain, ytest = tts(X, y, test_size=0.3, random_state=78)
```

```python
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(xtrain, ytrain)
```

```python
knn_pred = knn.predict(xtest)
```

```python
knn_score = knn.score(xtest,ytest)
knn_score
```

```python
knn_mae = mean_absolute_error(ytest, knn_pred)
knn_mae
```

```python
knn_mse = mean_squared_error(ytest, knn_pred)
knn_mse
```

# Data Frame

```python
df = pd.DataFrame({
    'Name':['Linear Regression', 'KNN'],
    'Accuracy Score':[lin_score, knn_score],
    'Mean Absolute Error': [lin_mae, knn_mae],
    'Mean Squared Error':[lin_mse, knn_mse]})
df.head()
```

```python
df.plot(x="Name", y=["Accuracy Score","Mean Absolute Error","Mean Squared Error"], kind="bar")
```

```python
columns = df.columns
```

```python
plt.figure(figsize=(24,6))

for i in range(1,4):
   plt.subplot(1, 3, i)
   sns.barplot(data=df, x="Name", y=columns[i])
   plt.xlabel("")
   plt.ylabel("")
   plt.title(columns[i])
```
